```python
# 导入numpy

import numpy as np
```

```python
x = np.array([1.0, 2.0, 3.0])
print(x)
print(type(x))
```

```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x / 2.0)
```

```python
A = np.array([[1, 2], [3, 4]])
print(A)
print(A.shape)
print(A.dtype)
```

```python
B = np.array([[3, 0], [0, 6]])
print(A + B)
print(A * B)
print(A * 10)
```

```python
x = np.array([[51, 55], [14, 19], [0, 4]])
print(x)
print(x[0])
print(x[0][1])
print(x[1:,1:])
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.savefig("./data/test.jpg")
plt.show()
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = "sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```

```python
from matplotlib.image import imread
img = imread("./data/test.jpg")
plt.imshow(img)
plt.show()
```
